# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def ReLU(x):
    y=list(map(lambda x:x if x>0 else 0,x))
    return y
    
x = plt.linspace(-10,10,100)

#ReLU函数图像
plt.plot(x, ReLU(x), 'b', label='linspace(-10,10,100)')
plt.grid()

# 顯現圖示的Title
plt.title('ReLU Function')

plt.show()
```

```python
#ReLU函数导数图像
def dReLU(x):
    dev_y =list(map(lambda x: 1 if x>0 else 0,x))
    return dev_y

plt.plot(x, dReLU(x), 'r', label='linspace(-10,10,100)')
plt.grid()

# 顯現圖示的Title
plt.title('dReLU Function')

plt.show()
```
